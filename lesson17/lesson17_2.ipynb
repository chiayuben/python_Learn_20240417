```python
df2 = df1.reindex(columns=['洲名','國家','日期','總確診數','新增確診數'])
df2.info
```

```python
df2 = df1.reindex(columns=['洲名','國家','日期','總確診數','新增確診數'],)
taiwan_df=df2.query('國家=="台灣" ')
taiwan_df
```

```python
tai2021_df=taiwan_df.query('日期>="2021-01-01" and 日期<="2021-12-31"')
tai2021_df
```

```python
tai2021_df.to_csv('taiwan2021.csv',index=False)
#可以將我們整理好的資料，轉成CSV給別人。
```
