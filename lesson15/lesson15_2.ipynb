```python
#較精簡的寫法
#如何寫成中文

'''先在linux系統安裝兩個功能
1.sudo apt-get update -y
2.sudo apt-get install -y fonts-wqy-zenhei


路徑:
#mac
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#windows
font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)


'''

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)


student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}

subject =list(student1.keys())
values1 =list(student1.values())
values2 = list(student2.values())

subject, values1, values2

#繪圖
import matplotlib.pyplot as plt

figure =plt.figure(figsize=(8,5),dpi=250,facecolor='#F19483')#顏色的部分可以上網找 日本色標
axes= figure.add_subplot(1,1,1)
axes.plot(subject,values1,'bo--',label="學生1")
axes.plot(subject,values2,'ro-.',label="學生2")
axes.grid(axis='y') #格線
axes.set_title("學生1 vs 學生2",fontproperties=font) #標題(只能英文)
axes.set_xlabel('科目',fontproperties=font) #X軸的標題
axes.set_ylabel('分數',fontproperties=font) #Y軸的標題
axes.legend([f'學生1',f'學生2'],prop=font)
print(axes.get_xticks())
print(axes.get_yticks())
labels=axes.get_xticklabels()
#print(labels)
for text in labels:
    subject_name=text.get_text()
    score1=student1[subject_name] #為了從Dict得到分數
    x_value = text.get_position()[0]
    axes.annotate(str(score1),(x_value-0.05,score1+0.5))
    score2 = student2[subject_name]
    axes.annotate(str(score2),(x_value-0.05,score2+0.5))


plt.show()
```
